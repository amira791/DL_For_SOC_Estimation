## CHARGEMENT et Exploration des DONN√âES

```python
#Importation des bibs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Chargement du dataset
dataset_path = r"C:\Users\VECTUS\DL_For_SOC_Estimation\datasets\concat_data.csv"

# Lecture du CSV
df = pd.read_csv(dataset_path, delimiter=",")

# V√©rification des colonnes disponibles
print("\U0001F4CA Colonnes d√©tect√©es :", df.columns)

# Aper√ßu des premi√®res lignes
display(df.head())
```

```python
# V√©rification des valeurs manquantes
print("\n V√©rification des valeurs manquantes :")
print(df.isnull().sum())

# Statistiques descriptives
display(df.describe())
```

### Description des colonnes du fichier CSV

| Nom de la colonne | Description |
|-------------------|-------------|
| **Heure** | Temps en secondes depuis le d√©but de la mesure |
| **Courant** | Valeur du courant de la batterie haute tension (en amp√®res) |
| **Tension** | Tension de la batterie haute tension (en volts) |
| **Cellule_min** | Tension minimale parmi les cellules de la batterie HV (en millivolts) |
| **Cellule_max** | Tension maximale parmi les cellules de la batterie HV (en millivolts) |
| **Temp1** | Temp√©rature relev√©e √† un premier point de la batterie HV (en ¬∞C) |
| **Temp4** | Temp√©rature relev√©e √† un quatri√®me point de la batterie HV (en ¬∞C) |
| **SOC** | √âtat de charge de la batterie en pourcentage |




```python
# === 2. VISUALISATIONS ===
plt.figure(figsize=(12, 6))
sns.histplot(df['SOC'], bins=30, kde=True, color='blue')
plt.title('Distribution du SOC')
plt.xlabel('State of Charge (SOC)')
plt.ylabel('Fr√©quence')
plt.grid()
plt.show()
```

```python
# Corr√©lation entre les variables
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de corr√©lation')
plt.show()
```

```python

# Scatter plot entre SOC et courant
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Tension'], y=df['SOC'], alpha=0.6)
plt.title('Relation entre le courant et le SOC')
plt.xlabel('Courant (A)')
plt.ylabel('SOC')
plt.grid()
plt.show()
```
